```python
uso_cpu = [
[20, 25, 40, 50, 45, 60, 55, 35, 70, 65],
[30, 35, 50, 60, 40, 55, 60, 45, 50, 55],
[15, 20, 30, 25, 35, 40, 45, 50, 55, 60],
[10, 15, 25, 35, 45, 50, 55, 60, 65, 70],
]

def uso_maximo(regiao):
  return max(regiao)

for i, regiao in enumerate(uso_cpu):
  print(f"Uso máximo na região {i+1}= {uso_maximo(regiao)}%")
```

```python
def acima_media(regiao):
    media = sum(regiao) / len(regiao)
    acima_media = sum(1 for valor in regiao if valor > media)
    return acima_media

for i, regiao in enumerate(uso_cpu):
    quantidade_acima_media = acima_media(regiao)
    print(f"Região {i+1} = {quantidade_acima_media} valores acima da média")
```

```python
def dados_decrescente(lista):
    return sorted(lista, reverse=True)

for i, regiao in enumerate(uso_cpu):
    ordem_decrescente_uso_cpu = dados_decrescente(regiao)
    print(f"Região {i+1} = {ordem_decrescente_uso_cpu} Com o maior para o menor")
```
